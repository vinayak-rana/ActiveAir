### Run the below cell to start the download process

```python
# Helper function for interacting with placeholders
def interact_with_placeholder(k):
    placeholders = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "placeholder")))
    placeholders[0].click()
    ul_element = wait.until(EC.presence_of_element_located((By.TAG_NAME, "ul")))
    list_items = ul_element.find_elements(By.TAG_NAME, "li")
    list_items[k].click()

def func():
    dropdown_button = driver.find_element(By.CLASS_NAME, "c-btn")
    dropdown_button.click()
    pm25_checkbox = driver.find_element(By.XPATH, "//label[text()='PM2.5']/preceding-sibling::input")
    pm10_checkbox = driver.find_element(By.XPATH, "//label[text()='PM10']/preceding-sibling::input")

    # Use ActionChains to move to the checkbox and click it
    actions = ActionChains(driver)

    # Click PM2.5 checkbox
    actions.move_to_element(pm25_checkbox).click().perform()

    # Click PM10 checkbox
    actions.move_to_element(pm10_checkbox).click().perform()

    dropdown_button.click()
    dropdowns = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "ng-select.select-box")))
    dropdowns[-1].click()
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.CSS_SELECTOR, "div.options ul"))
    )

    # Find the "30 Minute" option by text and click it
    thirty_min_option = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//li[normalize-space(text())='30 Minute']"))
    )
    driver.execute_script("arguments[0].scrollIntoView();", thirty_min_option)
    thirty_min_option.click()

    calendar_icon = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div.wc-date-container i.fa-calendar")))

    # Click the calendar icon to open the date picker
    calendar_icon.click()

    prev_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "i.wc-prev.fa.fa-angle-left"))
    )

    # Click the "Previous" button
    for _ in range(11):
        prev_button.click()

    day_element = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//td[@class='calendar-day' and .//span[text()='1']]"))
    )

    # Click the day element
    day_element.click()

    submit_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//button[@class='btn btn-primary' and text()='Submit']"))
    )

    # Click the "Submit" button
    submit_button.click()

    excel_link = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "a.btn.btn-success.tooltip-bottom[data-tooltip='Excel']"))
    )
    excel_link.click()

    time.sleep(15)  # Waitig time for dwnld

def run_with_captcha_handling():
    start_time = time.time()

    for i in range(31):
        driver.get('https://www.google.com')
        driver.get(f'{url}/data')

        interact_with_placeholder(i)

        for j in range(len(list_items)):
            driver.get('https://www.google.com')
            driver.get(f'{url}/data')

            interact_with_placeholder(i)

            for k in range(len(list_items)):
                driver.get('https://www.google.com')
                driver.get(f'{url}/data')

                interact_with_placeholder(i)
                interact_with_placeholder(j)
                interact_with_placeholder(k)
                func()

            driver.get('https://www.google.com')
            driver.get(f'{url}/data')

        # After 5 minutes, pause the script for CAPTCHA handling
        elapsed_time = time.time() - start_time
        if elapsed_time >= 900: #15 mins
            print("It's time to solve the CAPTCHA. The script will resume after 5 minutes.")
            driver.get('https://airquality.cpcb.gov.in/ccr/#/caaqm-dashboard-all/caaqm-landing')
            time.sleep(300)  # Pause for 5 minutes

run_with_captcha_handling()
```
